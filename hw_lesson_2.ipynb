Score Vectorizers

```python
# Загружаем данные
data = open('corpus').read()
labels, texts = [], []
for i, line in enumerate(data.split("\n")):
    content = line.split()
    labels.append(content[0])
    texts.append(" ".join(content[1:]))

# создаем df
trainDF = pd.DataFrame()
trainDF['text'] = texts
trainDF['label'] = labels
trainDF.head(5)
```

```python
from sklearn import model_selection, preprocessing, linear_model

train_x, valid_x, train_y, valid_y = model_selection.train_test_split(trainDF['text'], trainDF['label'])

# labelEncode целевую переменную
encoder = preprocessing.LabelEncoder()
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)


count_vect = CountVectorizer(ngram_range=(2, 2), analyzer='word', 
                                   max_df=1.0, max_features = 10000, stop_words='english')
count_vect.fit(trainDF['text'])

xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_count, train_y)
predictions = classifier.predict(xvalid_count)
predictions
```

```python
xtrain_count.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
count_score = accuracy_score(valid_y, predictions)
count_score
```

```python
vct = TfidfVectorizer(max_df=1.0, max_features = 10000, stop_words='english', token_pattern=r'\w{1,}')
vct.fit(trainDF['text'])

xtrain_tf =  vct.transform(train_x)
xvalid_tf =  vct.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_tf, train_y)
predictions = classifier.predict(xvalid_tf)
predictions
```

```python
tf_score = accuracy_score(valid_y, predictions)
tf_score
```

```python
h = HashingVectorizer(analyzer='word', n_features = 10000,stop_words='english')
h.fit(trainDF['text'])

xtrain_hash =  h.transform(train_x)
xvalid_hash =  h.transform(valid_x)

classifier = linear_model.LogisticRegression()
classifier.fit(xtrain_hash, train_y)
predictions = classifier.predict(xvalid_hash)
predictions
```

```python
hs = accuracy_score(valid_y, predictions)
hs
```

```python
df = pd.DataFrame(columns=['accuracy'])
```

```python
df.loc['CountVectorizer'] = [count_score]
df.loc['TfidfVectorizer'] = [tf_score]
df.loc['HashingVectorizer'] = [hs]
df
```

При изменении количества параметров в векторайзерах, а также при увеличении параметра n_features c 1000 до 10000 
и параметра max_df с 0.9 до 1.0 скоры растут у всех трёх векторайзеров.
